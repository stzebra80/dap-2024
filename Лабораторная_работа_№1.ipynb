## 1. Типы ячеек в Jupyter Notebook

В Jupyter Notebook существует **четыре основных типа ячеек**:

- **Code (код)** — для написания и выполнения кода Python или других языков;
- **Markdown** — для форматированного текста, заголовков, списков, ссылок и других элементов оформления;
- **Raw NBConvert** — для текста, который должен остаться без изменений при экспорте ноутбука;
- **Heading (устаревший)** — ранее использовался для заголовков, сейчас заменён на Markdown с синтаксисом заголовков.

> Переключение между типами ячеек осуществляется через меню **Cell → Cell Type** или с помощью клавиатурных сокращений: нажмите `M` для Markdown, `Y` для Code в командном режиме.

## 2. Как с помощью Markdown сформировать иерархический список?

Для создания иерархического (вложенного) списка в Markdown необходимо использовать **отступы**: каждый уровень вложенности обозначается двумя или четырьмя пробелами либо символом табуляции перед маркером элемента.

Пример кода:
* Элемент 1
  * Подэлемент 1.1
    * Подэлемент 1.1.1
* Элемент 2
  - Альтернативный маркер

  
> Главное — соблюдать единообразие отступов внутри всего списка.

## 3. Как с помощью Markdown оформить цитату?

Цитата в Markdown оформляется с помощью символа **больше** (`>`) в начале строки.

> Это обычная цитата.
>
> Это второй абзац той же цитаты.
>> А это вложенная цитата внутри основной.
> Символы `>` видны как обычный текст внутри серого блока кода.



## 4. Как в текстовую ячейку вставить отформатированный блок кода?

Для вставки отформатированного блока кода в Markdown-ячейке используется **тройной обратный апостроф** (```) в начале и конце блока.

Пример кода:
```
import pandas as pd
df = pd.read_csv('data.csv')
print(df.head())
```

> Для вставки кода в одну строку внутри текста используются одинарные обратные апострофы: `код`.

## 5. Сколько уровней заголовка существует в Markdown?
В Markdown предусмотрено **шесть уровней заголовков**, которые обозначаются символами решётки (`#`) от одного до шести:

| Синтаксис | Уровень | Пример отображения |
|-----------|---------|-------------------|
| `#` | H1 | Самый крупный заголовок |
| `##` | H2 | Подзаголовок раздела |
| `###` | H3 | Подраздел |
| `####` | H4 | Детализация |
| `#####` | H5 | Малозаметный заголовок |
| `######` | H6 | Минимальный уровень |

> Альтернативный синтаксис для первых двух уровней — подчёркивание текста символами `=` (для H1) или `-` (для H2) на следующей строке, но он используется реже.

## 6. Шаги процесса анализа данных (подробно про очистку и предобработку)
Процесс анализа данных обычно включает следующие этапы:

1. Постановка задачи

### Подробно про очистку и предобработку данных

Это критически важный этап, поскольку качество результатов напрямую зависит от качества входных данных.

**Основные задачи этапа:**

- **Обработка пропущенных значений**: удаление строк, импутация средним/медианой/модой или с помощью моделей;
- **Выявление и обработка выбросов**: методы IQR, Z-score, визуальный анализ;
- **Приведение типов данных** к корректному формату;
- **Кодирование категориальных признаков**: One-Hot Encoding, Label Encoding;
- **Нормализация или стандартизация** числовых признаков;
- **Устранение дубликатов** и несоответствий;
- **Проверка на мультиколлинеарность**;
- **Балансировка классов** в задачах классификации.

> **Инструменты**: `pandas`, `numpy`, `scikit-learn` в Python. Без качественной предобработки даже сложные модели могут давать некорректные или смещённые прогнозы.

## 7. Типы данных (подробно про порядковые категориальные данные)

В анализе данных встречаются следующие основные типы:

- Числовые (непрерывные и дискретные)
- Категориальные (номинальные и порядковые) ← рассматриваем подробно
- Текстовые
- Временные ряды
- Геопространственные
- Бинарные

### Порядковые категориальные данные

Это тип данных, значения которого имеют **естественный порядок**, но не имеют фиксированного числового интервала между категориями.

**Примеры:**

| Категория | Примеры значений |
|-----------|-----------------|
| Образование | школа < бакалавр < магистр < доктор |
| Удовлетворённость | плохо < нормально < хорошо < отлично |
| Размеры одежды | S < M < L < XL |

**Особенности работы:**

- При кодировании целесообразно использовать **Ordinal Encoding**, присваивая категориям возрастающие целые числа — это позволяет алгоритмам машинного обучения учитывать иерархию;
- Необходимо помнить, что математические операции над такими закодированными значениями (например, среднее) могут быть неинтерпретируемы;
- Визуализация часто осуществляется с помощью столбчатых диаграмм с упорядоченными категориями или коробчатых диаграмм при сравнении с числовыми признаками.

## 8. Что такое Jupyter Notebook, для чего используется?

**Jupyter Notebook** — это интерактивная веб-среда для создания и совместного использования документов, содержащих:

- ✅ Исполняемый код
- ✅ Визуализации и графики
- ✅ Форматированный текст (Markdown)
- ✅ Математические формулы (LaTeX)

**История названия**: Ju-py-ter = **Ju**lia + **Py**thon + **R** (поддержка множества языков).

**Основные сценарии использования:**

1. Разведочный анализ данных (EDA)
2. Прототипирование моделей машинного обучения
3. Обучение и демонстрация программирования
4. Создание интерактивных отчётов и дашбордов
5. Научные исследования с воспроизводимыми вычислениями

> Поддерживает экспорт в HTML, PDF, Markdown, слайды. Интегрируется с Git, JupyterHub для командной работы, облачными платформами.

## 10. Обязательно ли использовать Jupyter Notebook для анализа данных?

**Нет, не обязательно.** Jupyter Notebook — это один из удобных инструментов, но не единственный.

**Альтернативные варианты:**

| Инструмент | Лучше подходит для |
|------------|-------------------|
| PyCharm, VS Code | Промышленная разработка, большие проекты |
| Обычные .py-скрипты | Автоматизация, CI/CD-пайплайны |
| RStudio | Анализ на языке R |
| KNIME, RapidMiner | Визуальное программирование без кода |
| Excel / Google Sheets | Простые задачи, быстрый просмотр |

**Когда выбирать Jupyter/Colab:**
- ✅ Интерактивное исследование данных
- ✅ Обучение и демонстрация
- ✅ Быстрое прототипирование
- ✅ Создание отчётов с кодом и визуализациями

**Когда выбирать другие инструменты:**
- ✅ Продакшен-разработка
- ✅ Командная работа с жёстким контролем версий
- ✅ Автоматизация регулярных задач

> Главное — не инструмент, а методология: корректная постановка задачи, качество данных, воспроизводимость результатов и интерпретируемость выводов.
