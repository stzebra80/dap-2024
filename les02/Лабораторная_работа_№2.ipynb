Задача №1. https://leetcode.com/problems/two-sum/

```python
class Solution1(object):
    def twoSum(nums, target):

        num_to_index = {}

        for i, num in enumerate(nums):
            complement = target - num

            if complement in num_to_index:
                return [num_to_index[complement], i]

            num_to_index[num] = i

        return []

    print(twoSum([2,7,11,15], 9))
    print(twoSum([3,2,4], 6))
    print(twoSum([3,3], 6))
```

Задача №3. https://leetcode.com/problems/roman-to-integer/description/

```python
def romanToInt(s: str) -> int:
    roman_map = {
        'I': 1, 'V': 5, 'X': 10, 'L': 50,
        'C': 100, 'D': 500, 'M': 1000
    }

    total = 0
    n = len(s)

    for i in range(n):
        value = roman_map[s[i]]

        if i < n - 1 and value < roman_map[s[i + 1]]:
            total -= value
        else:
            total += value

    return total

print(romanToInt("III"))      # 3
print(romanToInt("IV"))       # 4
print(romanToInt("IX"))       # 9
print(romanToInt("XII"))      # 12
print(romanToInt("XXVII"))    # 27
print(romanToInt("MCMXCIV"))  # 1994
```

Задача №4. https://leetcode.com/problems/longest-common-prefix/description/

```python
class Solution(object):
    def longestCommonPrefix(strs):
        if not strs:
            return ""

        prefix = strs[0]

        for s in strs[1:]:
            while not s.startswith(prefix):
                prefix = prefix[:-1]
                if not prefix:
                    return "No common prefix"

        return prefix


    words = ["flower","flow","flight"]
    print(longestCommonPrefix(words))
    words = ['flower', 'car', 'legs']
    print(longestCommonPrefix(words))
```

Задача №5. https://leetcode.com/problems/merge-two-sorted-lists/description/

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def mergeTwoLists(self, list1, list2):
        if list1.val <= list2.val:
            head = list1
            list1 = list1.next
        else:
            head = list2
            list2 = list2.next

        current = head

        while list1 and list2:
            if list1.val <= list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next
            current = current.next

        if list1:
            current.next = list1
        else:
            current.next = list2

        return head

def createList(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

def listToArray(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result

# Тестовые данные

test_cases = [
    {"list1": [1, 2, 4],      "list2": [1, 3, 4],    "expected": [1, 1, 2, 3, 4, 4]},
    {"list1": [1, 3, 5],      "list2": [2, 4, 6],    "expected": [1, 2, 3, 4, 5, 6]},
    {"list1": [1, 2, 3],      "list2": [4, 5, 6],    "expected": [1, 2, 3, 4, 5, 6]},
    {"list1": [4, 5, 6],      "list2": [1, 2, 3],    "expected": [1, 2, 3, 4, 5, 6]},
    {"list1": [1],            "list2": [2],          "expected": [1, 2]},
    {"list1": [2],            "list2": [1],          "expected": [1, 2]},
    {"list1": [1, 1, 1],      "list2": [1, 1, 1],    "expected": [1, 1, 1, 1, 1, 1]},
    {"list1": [-5, -2, 0],    "list2": [-3, 1, 3],   "expected": [-5, -3, -2, 0, 1, 3]},
    {"list1": [1, 3, 5, 7],   "list2": [2, 4],       "expected": [1, 2, 3, 4, 5, 7]},
    {"list1": [2, 4],         "list2": [1, 3, 5, 7], "expected": [1, 2, 3, 4, 5, 7]},
]

# Запуск тестов

print("=" * 60)
print("ТЕСТИРОВАНИЕ: Merge Two Sorted Lists")
print("=" * 60)

solution = Solution()
passed = 0
failed = 0

for i, test in enumerate(test_cases, 1):
    list1 = createList(test["list1"])
    list2 = createList(test["list2"])


    merged = solution.mergeTwoLists(list1, list2)
    result = listToArray(merged)

    if result == test["expected"]:
        status = "PASS"
        passed += 1

    print(f"\nТест {i}:")
    print(f"  list1:  {test['list1']}")
    print(f"  list2:  {test['list2']}")
    print(f"  Ожидается: {test['expected']}")
    print(f"  Получено:  {result}")
    print(f"  Статус: {status}")

```

Задача №6. https://leetcode.com/problems/longest-substring-without-repeating-characters/description/

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        char_set = set()
        left = 0
        max_len = 0

        for right in range(len(s)):
            while s[right] in char_set:
                char_set.remove(s[left])
                left += 1
            char_set.add(s[right])
            max_len = max(max_len, right - left + 1)

        return max_len

test_cases = [
    "abcabcbb",
    "bbbbb",
    "pwwkew",
    "",
    "a",
    "au",
    "dvdf",
    "anviaj",
    "tmmzuxt",
    "abcde",
    "abba",
    "aab",
    "abcdc",
    " ",
    "   ",
    "abcdcdef",
]

solution = Solution()

for s in test_cases:
    result = solution.lengthOfLongestSubstring(s)
    print(f"Вход: '{s}' -> Результат: {result}")
```

## Вопросы

1. Что такое динамическая типизация и чем она отличается от статической? Динамическая типизация означает, что тип переменной определяется во время выполнения программы и может изменяться в ходе работы, тогда как при статической типизации тип фиксируется на этапе компиляции и изменение типа переменной требует явного приведения или запрещено.

2. Чем отличается цикл for в Python от цикла for в С++? В Python цикл for перебирает элементы итерируемого объекта напрямую без использования индексов, а в С++ он чаще всего работает со счетчиком или итераторами, требуя явного указания условия продолжения и шага инкремента в заголовке цикла.

3. Для чего предназначены операторы continue и break? Оператор break полностью завершает выполнение цикла и передает управление следующей после цикла инструкции, а оператор continue пропускает оставшуюся часть текущей итерации и сразу переходит к следующей проверке условия цикла.

4. Как определить функцию в python? Может ли функция возвращать несколько значений? Функция определяется ключевым словом def, за которым следует имя и параметры в скобках. Функция может возвращать несколько значений, фактически возвращая кортеж, который автоматически распаковывается при присваивании нескольким переменным.

5. Что такое лямбда функции и для чего они предназначены? Лямбда функции это анонимные функции, создаваемые с помощью ключевого слова lambda в одну строку. Они предназначены для написания коротких выражений в местах, где создание полноценной именованной функции избыточно, например в качестве аргумента для других функций.

6. Может ли тело функции содержать определение другой функции? Тело функции может содержать определение другой функции, что часто используется для создания замыканий или инкапсуляции вспомогательной логики, которая не должна быть видна снаружи.

7. Как определить класс в python? Для чего нужна переменная self? Класс определяется ключевым словом class, за которым следует имя класса. Переменная self ссылается на текущий экземпляр класса и необходима для доступа к его атрибутам и методам внутри методов класса.

8. Как определить приватный метод? Можно ли вызвать приватный метод за пределами класса? Приватный метод определяется добавлением двух подчеркиваний перед именем метода. Вызвать его за пределами класса технически возможно через механизм искажения имен, но это нарушает инкапсуляцию и не рекомендуется разработчиками.

9. Как определить статический метод класса? Для чего может понадобиться статический метод? Статический метод определяется декоратором @staticmethod перед объявлением метода. Он может понадобиться для создания utility функций, которые логически относятся к классу, но не требуют доступа к экземпляру или самому классу.

10. Чем staticmethod отличается от classmethod? Staticmethod не получает никаких неявных первых аргументов, а classmethod автоматически получает класс в качестве первого аргумента, что позволяет работать с классовыми переменными и наследованием.

11. Как определить собственный класс исключений Python? Как выбросить и перехватить исключение? Собственный класс исключений создается наследованием от базового класса Exception. Исключение выбрасывается оператором raise и перехватывается блоком try except с указанием типа исключения.

12. Что такое картеж, для чего он используется? Кортеж это неизменяемая последовательность элементов, которая создается с помощью круглых скобок. Он используется для хранения фиксированных данных, в качестве ключей словаря или когда требуется гарантия неизменности коллекции.
