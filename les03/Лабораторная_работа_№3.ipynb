2. Создать 8x8 матрицу и заполнить её единицами и нулями в шахматном порядке

````python
import numpy as np
matrix = np.zeros((8, 8), dtype=int)
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:
            matrix[i, j] = 1
print("8x8 матрица в шахматном порядке:")
print(matrix)
````

3. Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8

4. Найти среднее заначение строки матрицы и вычесть его из каждого елемента строки

````python
import numpy as np

matrix = np.random.randint(1, 10, size=(5, 5))

print("Исходная матрица:")
print(matrix)

row_means = matrix.mean(axis=1, keepdims=True)

centered_matrix = matrix - row_means

print("\nСредние значения строк:")
print(row_means)

print("\nМатрица после вычитания среднего:")
print(centered_matrix)
````

6. Дан массив 10x2 подеставлющий собой точки в декартовой системе координат,
преобразовать в полярную

---


````python
import numpy as np
cartesian = np.random.randint(-10, 11, size=(10, 2)).astype(float)
print("Декартовы координаты (x, y):")
print(cartesian)
r = np.sqrt(cartesian[:, 0]**2 + cartesian[:, 1]**2)
theta = np.arctan2(cartesian[:, 1], cartesian[:, 0])
polar = np.column_stack((r, theta))
print("\nПолярные координаты (r, θ в радианах):")
print(polar)
theta_deg = np.degrees(theta)
polar_deg = np.column_stack((r, theta_deg))
print("\nПолярные координаты (r, θ в градусах):")
print(polar_deg)
````

8. Дан массив, добавить 1 к каждому элементу с индексом, заданным в другом массиве. Добавление производиться только 1 раз

````python
import numpy as np

array = np.random.randint(0, 101, size=10)

indices = np.random.randint(0, len(array), size=15)

print("Исходный массив:")
print(array)
print("\nИндексы для изменения:")
print(indices)

unique_indices = np.unique(indices)
print("\nУникальные индексы:")
print(unique_indices)

array[unique_indices] += 1

print("\nМассив после изменения:")
print(array)

print("\nИзменённые элементы:")
for idx in unique_indices:
    print(f"  Индекс {idx}: значение увеличено на 1")
````

9. Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением

````python
import numpy as np

z = np.array([1, 2, 3, 4, 5])

print("Исходный вектор z:")
print(z)
k = 3
step = k + 1
new_length = len(z) + (len(z) - 1) * k
new_z = np.zeros(new_length)
new_z[::step] = z
print("\nНовый вектор с тремя нулями между каждым значением:")
print(new_z)
````

10. Найти ближайшее к заданному значению число в заданном массиве

````python
import numpy as np
z = np.arange(100)
v = np.random.uniform(0, 100)
print("Массив z:")
print(z)
print(f"\nЗаданное значение v: {v}")
differences = np.abs(z - v)
closest_index = np.argmin(differences)
closest_value = z[closest_index]
print(f"\nБлижайшее значение: {closest_value} (индекс {closest_index})")
````

**5**. Дан двумерный массив grid, в чем различее между операциями?


``` python
grid = np.arange(16).reshape((4, 4))
```

``` python
np.split(grid, [2])
np.split(grid, [2], axis=1)
```
```
Различие заключается в оси, вдоль которой выполняется разделение массива.
В первом случае np.split(grid, [2]) параметр axis не указан, поэтому
используется значение по умолчанию axis=0, что означает разделение массива по
строкам. Индекс 2 указывает на то, что разрыв произойдет после второй строки, в
результате чего функция вернет список из двух массивов формой (2, 4), где
первый массив содержит строки с индексами 0 и 1, а второй массив содержит
строки с индексами 2 и 3. Во втором случае np.split(grid, [2], axis=1) явно
указана ось axis=1, что означает разделение массива по столбцам. Индекс 2
указывает на разрыв после второго столбца, в результате чего функция вернет
список из двух массивов формой (4, 2), где первый массив содержит столбцы с
индексами 0 и 1, а второй массив содержит столбцы с индексами 2 и 3. Таким
образом, первая операция делит матрицу горизонтально на две части по строкам, а
вторая операция делит матрицу вертикально на две части по столбцам.
```


**6**. В чем различие между операциями?

``` python
    [1, 2, 3] + [3, 2, 1]
    np.array([1, 2, 3]) + np.array([3, 2, 1])
```
```
Первая операция выполняет конкатенацию стандартных списков Python, соединяя их
в один список длиной шесть элементов [1, 2, 3, 3, 2, 1]. Вторая операция
выполняет поэлементное арифметическое сложение массивов NumPy, складывая
соответствующие значения с одинаковыми индексами, что дает результат array([4,
4, 4]). Различие заключается в перегрузке оператора плюс для разных типов
данных: для list это объединение последовательностей, а для ndarray это
векторизованная математическая операция.
```

**7**. Как в CodeLab измерить время выполнения кода?
```
В среде выполнения на базе Jupyter Notebook, к которой относится Google Colab,
для измерения времени выполнения кода чаще всего используют магические команды
IPython. Команда %time измеряет время однократного выполнения ячейки или строки
кода и выводит время процессора и реальное время выполнения, а для измерения

быстрого кода, используется команда %timeit, которая автоматически запускает
код несколько раз и выводит среднее время выполнения с отклонением. Также можно
использовать стандартный модуль Python time, замеряя разницу между временем до
и после выполнения кода, или модуль timeit для более строгих бенчмарков. В
настройках Google Colab можно включить отображение времени выполнения каждой
ячейки через меню Инструменты и раздел Настройки, где активируется опция показа
времени выполнения ячеек.
```

**10**.  В чем преимущество массивов NumPy по сравнению с обычными списками.  Чем обусловлено это преимущество?
```
Главное преимущество массивов NumPy перед обычными списками Python заключается
в значительно более высокой скорости выполнения вычислений и эффективности
использования памяти.
Это преимущество обусловлено следующими факторами:
Однородность данных (Homogeneity): В отличие от списков, которые могут хранить
объекты разных типов, массивы NumPy хранят элементы только одного типа данных
(например, только int32 или float64). Это позволяет хранить данные в
непрерывном блоке памяти без накладных расходов на хранение информации о типе
каждого отдельного элемента, что характерно для объектов Python в списках.
Векторизация (Vectorization): Операции в NumPy выполняются над всем массивом
целиком на уровне компилированного кода (C/C++), избегая медленных циклов
интерпретатора Python. Это устраняет необходимость писать явные циклы for для
перебора элементов.
Локальность данных и кэш-память: Благодаря хранению данных в непрерывном
участке памяти, процессор может эффективнее использовать кэш-память при
обработке массивов, что ускоряет доступ к данным по сравнению со списками, где
элементы могут быть разбросаны в памяти.
Оптимизированные алгоритмы: Библиотека NumPy использует высокооптимизированные
математические библиотеки (такие как BLAS и LAPACK), написанные на
низкоуровневых языках, что обеспечивает максимальную производительность для
линейной алгебры и статистических операций.
```
