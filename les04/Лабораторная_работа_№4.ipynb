8.  Найдите самое популярное имя среди пассажиров Титаника мужского пола?

````python
import re
from collections import Counter

df = pd.read_csv('titanic.csv')

men_df = df[df['Sex'] == 'male']

def extract_name(full_name):
    try:
        part_after_comma = full_name.split(',')[1]
        words = part_after_comma.strip().split()

        for word in words:
            clean_word = word.replace('.', '')
            if clean_word not in ['Mr', 'Master', 'Dr', 'Rev', 'Col', 'Major', 'Sir', 'Jonkheer']:
                return clean_word
    except IndexError:
        return None

men_df['FirstName'] = men_df['Name'].apply(extract_name)

name_counts = Counter(men_df['FirstName'])
most_common_name, count = name_counts.most_common(1)[0]

print(f"Самое популярное имя среди мужчин: {most_common_name}")
print(f"Количество повторений: {count}")
print("\nТоп-5 популярных имен:")
for name, count in name_counts.most_common(5):
    print(f"{name}: {count}")
````

Дисперсия — это мера разброса данных, показывающая, насколько сильно значения
отклоняются от их среднего арифметического. Формально это средний квадрат
отклонений значений от их математического ожидания. Чем выше дисперсия, тем

3. Что такое среднеквадратичное отклонение? Рассчитать среднеквадратичное отклонение для предложенного ряда чисел.
```
Среднеквадратичное отклонение (СКО или стандартное отклонение) — это корень

```
4. Что такое Series и чем он отличается от DataFrame
```
Что такое Series и чем он отличается от DataFrame?
Series в pandas — это одномерный массив данных с подписанными индексами
(метками). По сути, это один столбец таблицы.
DataFrame — это двумерная таблица данных (строки и столбцы), где каждый столбец
является объектом Series.
Главное отличие: Series хранит одну последовательность значений (один признак),
а DataFrame — коллекцию таких последовательностей, объединенных общими
индексами (множество признаков).
```

6. Можно ли выполнить операцию сложения над двумя объектами Series, если они имеют:
- одинаковые типы данных, одинаковые размеры и одинаковые индексы?
- одинаковые типы данных, одинаковые размеры, но разные индексы?,  
- одинаковые типы данных, разные размеры и одинаковые индексы?
- разные типы данных, одинаковые размеры и одинаковые индексы?
- разные типы данных, одинаковые размеры и разные индексы?
- разные типы данных, разные размеры и разные индексы?
```
Можно ли выполнить операцию сложения над двумя объектами Series...

Операция сложения в pandas выполняется по индексам (выравнивание), а не просто
по позициям.

Одинаковые типы, размеры и индексы: Да, сложится поэлементно. Результат будет
корректным Series той же длины.

Одинаковые типы, размеры, но разные индексы: Да, выполнится, но результат будет
содержать NaN там, где индексы не совпали. Если индексы полностью различны,
результат будет состоять только из NaN (длина суммы уникальных индексов).

Одинаковые типы, разные размеры, одинаковые индексы: Технически невозможно
иметь "одинаковые индексы" при разных размерах, если под этим подразумевается
полный набор меток. Если один индекс является подмножеством другого, сложение
пройдет по пересекающимся меткам, а для отсутствующих в меньшем Series будут
NaN.

Разные типы данных, одинаковые размеры и индексы: Да, если типы совместимы
(например, int и float). pandas попытается привести типы к общему (обычно к
более широкому, например, float). Если типы несовместимы (например, string и
int без явного преобразования), возникнет ошибка.

Разные типы, размеры и индексы: Выполнится попытка сложения по пересекающимся
индексам с приведением типов. Результат скорее всего будет содержать много NaN
и может иметь тип object или float, в зависимости от данных.

Разные типы, размеры и индексы (полностью несовпадающие): Операция выполнится,
но результат будет Series, заполненный NaN, так как пересечения индексов нет.
Тип данных результата зависит от логики приведения типов pandas для операндов.
```

9. Чем отличается метод iloc от метода loс
```
loc (location) работает с метками индексов и именами столбцов. Срезы включают
правую границу (например, loc[1:5] включит строки с индексами 1, 2, 3, 4, 5).
iloc (integer location) работает с целочисленными позициями (как в обычных
списках Python). Срезы не включают правую границу (например, iloc[1:5] включит
строки с позициями 1, 2, 3, 4).
```
10. Можно ли добавить в DateFrame дополнительный Series?
```
Да, можно. Если присвоить новый Series существующему DataFrame как новый
столбец (например, df['new_col'] = series), pandas автоматически выровняет
данные по индексам. Если индексы Series совпадают с индексами DataFrame, данные
встанут корректно. Если не совпадают — появятся NaN.
```
